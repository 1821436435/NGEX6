Some typical NumPy 'random' functions:

```python
import numpy as np

# To generate a random float values between 0 and 1 (exclusive)
print(np.random.rand(3, 2)) # Creates a 3 x 2 matrix of random float values between 0 and 1 (exclusive)

# To generate random float values from standard normal distribution (mean = 0 and std = 1)
print(np.random.randn(5)) # Generates 5 random float values from standard normal distribution

# To generate random integer values within a range of values
print(np.random.randint(1, 100, 10)) # Generates 10 random integer values between 1 and 100 (exclusive)

# To randomly select an element from a given list of elements
print(np.random.choice([1, 2, 3, 4, 5, 6, 7, 8, 9], size=3)) # Three random numbers from the given array will be chosen

# To generate a random sample of values from a normally distributed with a specified mean and standard deviation
print(np.random.normal(loc=0, scale=1, size=10)) # 10 random values will be generated from a normally distributed data where 0 (loc) is the mean and 1 (scale) is the standard deviation

# To set a seed for a random number generation to ensure you get the same results later as well
print(np.random.seed(42)) # This ensures that any random numbers generated in the program are same every time you run the program. The seed value '42' can be any other value
```

AIM #1: Generate a very large dataset
1. Generate a dataset of 1 million random data items between 1 and 100 items using only pandas
2. Generate a dataset of 1 million random data items between 1 and 100 using only NumPy
3. Calculate the time it takes for both the above operations. 
    3.1. Import the 'time' module, and use the time() function to calculate current time
    3.2. Which one is faster and why?

```python
# Write your code for AIM #1 here
import numpy as np
import pandas as pd
import time

# Using pandas
start_time_pd = time.time()
pd_data = pd.DataFrame(np.random.randint(1, 100, size=(1000000, 1)), columns=['RandomInts'])
pandas_time = time.time() - start_time_pd

# Using NumPy
start_time_np = time.time()
np_data = np.random.randint(1, 100, size=1000000)
numpy_time = time.time() - start_time_np

print(f"Pandas time: {pandas_time:.6f} seconds")
print(f"NumPy time: {numpy_time:.6f} seconds")

# Determine which is faster
if pandas_time < numpy_time:
    print("Pandas is faster.")
else:
    print("NumPy is faster.")
```

AIM #2: Basic statistics
For the given dataset on sleep health and lifestyle, do the following
1. Using only pandas, load the dataset, calculate mean 'Sleep Duration', 'Systolic Blood Pressure', 'Diastolic Blood Pressure', 'Heart Rate' and 'Daily Steps'.
2. Do the same as in Step 1 using only NumPy
3. Using only pandas, first calculate correlation (across only the numerical variables), and then separate correlation between...
    Sleep duration and Age
    Sleep duration and Heart rate
    Sleep duration and Daily steps
4. Using only NumPy, do the same as Step 3
5. Using pandas only, calculate standard deviation for 'Sleep Duration'. 
6. Usiong NumPy only, calculate standard deviation for 'Sleep Duration'. 
7. Calculate the time difference between using pandas and NumPy, right from the step of loading the dataset to the final standard deviation step. 
    5.1. Which one is faster and why?

```python

import pandas as pd
import numpy as np
import time

# Load the dataset using pandas
start_time_pd = time.time()
df = pd.read_csv('sleep_health.csv')

# Calculate mean values
mean_values_pd = df[['Sleep Duration', 'Systolic Blood Pressure', 'Diastolic Blood Pressure', 'Heart Rate', 'Daily Steps']].mean()
print("Pandas Mean Values:\n", mean_values_pd)

# Calculate correlation
correlation_pd = df.corr()
sleep_age_corr = correlation_pd.loc['Sleep Duration', 'Age']
sleep_heart_rate_corr = correlation_pd.loc['Sleep Duration', 'Heart Rate']
sleep_daily_steps_corr = correlation_pd.loc['Sleep Duration', 'Daily Steps']
print(f"Correlation (Pandas) - Sleep Duration and Age: {sleep_age_corr}")
print(f"Correlation (Pandas) - Sleep Duration and Heart Rate: {sleep_heart_rate_corr}")
print(f"Correlation (Pandas) - Sleep Duration and Daily Steps: {sleep_daily_steps_corr}")

# Load the dataset using NumPy
data = np.genfromtxt('sleep_health.csv', delimiter=',', skip_header=1, dtype=None, encoding='utf-8', names=True)

# Calculate mean values using NumPy
mean_sleep_duration = np.mean(data['Sleep_Duration'])
mean_systolic_bp = np.mean(data['Systolic_Blood_Pressure'])
mean_diastolic_bp = np.mean(data['Diastolic_Blood_Pressure'])
mean_heart_rate = np.mean(data['Heart_Rate'])
mean_daily_steps = np.mean(data['Daily_Steps'])
print("NumPy Mean Values:\n", mean_sleep_duration, mean_systolic_bp, mean_diastolic_bp, mean_heart_rate, mean_daily_steps)

# Calculate correlation using NumPy
sleep_duration = data['Sleep_Duration']
age = data['Age']
heart_rate = data['Heart_Rate']
daily_steps = data['Daily_Steps']

sleep_age_corr_np = np.corrcoef(sleep_duration, age)[0, 1]
sleep_heart_rate_corr_np = np.corrcoef(sleep_duration, heart_rate)[0, 1]
sleep_daily_steps_corr_np = np.corrcoef(sleep_duration, daily_steps)[0, 1]
print(f"Correlation (NumPy) - Sleep Duration and Age: {sleep_age_corr_np}")
print(f"Correlation (NumPy) - Sleep Duration and Heart Rate: {sleep_heart_rate_corr_np}")
print(f"Correlation (NumPy) - Sleep Duration and Daily Steps: {sleep_daily_steps_corr_np}")

# Standard deviation using pandas
std_sleep_duration_pd = df['Sleep Duration'].std()
print(f"Standard Deviation (Pandas): {std_sleep_duration_pd}")

# Standard deviation using NumPy
std_sleep_duration_np = np.std(data['Sleep_Duration'])
print(f"Standard Deviation (NumPy): {std_sleep_duration_np}")

# Time difference
end_time_pd = time.time()
time_pd = end_time_pd - start_time_pd

start_time_np = time.time()
data_np = np.genfromtxt('sleep_health.csv', delimiter=',', skip_header=1, dtype=None, encoding='utf-8', names=True)
# Repeat calculations for NumPy as done previously...
end_time_np = time.time()
time_np = end_time_np - start_time_np

print(f"Total Time (Pandas): {time_pd:.6f} seconds")
print(f"Total Time (NumPy): {time_np:.6f} seconds")

# Determine which is faster
if time_pd < time_np:
    print("Pandas is faster.")
else:
    print("NumPy is faster.")
```

    5.4. Sleep Duration and Daily Steps

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import time

# Load the dataset using pandas
df = pd.read_csv('sleep_health.csv')

# Step 1: Using only pandas to plot distributions
plt.figure(figsize=(12, 10))

for i, column in enumerate(['Age', 'Sleep Duration', 'Quality of Sleep', 'Physical Activity Level', 'Stress Level', 'Heart Rate'], 1):
    plt.subplot(3, 2, i)
    sns.histplot(df[column], bins=30, kde=True)
    plt.title(f'Distribution of {column}')

plt.tight_layout()
plt.show()

# Step 2: Using NumPy for plotting distributions
data = np.genfromtxt('sleep_health.csv', delimiter=',', skip_header=1, dtype=None, encoding='utf-8', names=True)

plt.figure(figsize=(12, 10))

for i, column in enumerate(['Age', 'Sleep_Duration', 'Quality_of_Sleep', 'Physical_Activity_Level', 'Stress_Level', 'Heart_Rate'], 1):
    plt.subplot(3, 2, i)
    plt.hist(data[column], bins=30, alpha=0.7)
    plt.title(f'Distribution of {column}')

plt.tight_layout()
plt.show()

# Step 3: Using pandas for distribution based on categories
plt.figure(figsize=(12, 12))

for i, column in enumerate(['Quality of Sleep', 'Stress Level', 'Physical Activity Level', 'Occupation', 'BMI'], 1):
    plt.subplot(3, 2, i)
    sns.boxplot(x=column, y='Sleep Duration', data=df)
    plt.title(f'Sleep Duration by {column}')

plt.tight_layout()
plt.show()

# Step 4: Using NumPy for distribution based on categories
plt.figure(figsize=(12, 12))

for i, column in enumerate(['Quality_of_Sleep', 'Stress_Level', 'Physical_Activity_Level', 'Occupation', 'BMI'], 1):
    plt.subplot(3, 2, i)
    unique_values = np.unique(data[column])
    for value in unique_values:
        plt.hist(data['Sleep_Duration'][data[column] == value], bins=30, alpha=0.5, label=str(value))
    plt.title(f'Sleep Duration by {column}')
    plt.legend()

plt.tight_layout()
plt.show()

# Step 5: Using pandas for relationships
plt.figure(figsize=(12, 10))

for i, (x_col, y_col) in enumerate([('Age', 'Sleep Duration'), ('Sleep Duration', 'Heart Rate'), ('Heart Rate', 'Daily Steps'), ('Sleep Duration', 'Daily Steps')], 1):
    plt.subplot(2, 2, i)
    sns.scatterplot(x=df[x_col], y=df[y_col])
    plt.title(f'{x_col} vs {y_col}')

plt.tight_layout()
plt.show()

# Step 6: Using NumPy for relationships
plt.figure(figsize=(12, 10))

for i, (x_col, y_col) in enumerate([('Age', 'Sleep_Duration'), ('Sleep_Duration', 'Heart_Rate'), ('Heart_Rate', 'Daily_Steps'), ('Sleep_Duration', 'Daily_Steps')], 1):
    plt.subplot(2, 2, i)
    plt.scatter(data[x_col], data[y_col])
    plt.title(f'{x_col} vs {y_col}')

plt.tight_layout()
plt.show()

# Step 7: Time differences
start_time_pd = time.time()
# Include pandas plotting code here
end_time_pd = time.time()
time_pd = end_time_pd - start_time_pd

start_time_np = time.time()
# Include NumPy plotting code here
end_time_np = time.time()
time_np = end_time_np - start_time_np

print(f"Total Time (Pandas): {time_pd:.6f} seconds")
print(f"Total Time (NumPy): {time_np:.6f} seconds")

if time_pd < time_np:
    print("Pandas is faster.")
else:
    print("NumPy is faster.")
```
